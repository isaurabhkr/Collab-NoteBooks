```python
colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'black', 'cadetblue',
            'darkpurple', 'white', 'darkgreen', 'lightblue', 'lightgreen', 'darkblue', 'lightgray']
```

```python
def plot_points_on_map(map_obj, stops_df):
    for row in stops_df.itertuples():
        coord = (row.stop_lat, row.stop_lon)
        stop_name = row.stop_name
        folium.CircleMarker(location=coord,
                            radius=3,
                            fill=True,  # Set fill to True
                            tooltip=stop_name,
                            # color="black",
                            color=random.choice(colors),
                            fill_opacity=0.7).add_to(map_obj)
    print('plotted points')
```

### plotting stops based on some condition

```python
def conditionally_plot_points_on_map(map_obj, stops_df):
    for row in stops_df.itertuples():
        coord = (row.stop_lat, row.stop_lon)
        stop_name = row.stop_name
        if stop_name.lower().__contains__("aiims"):
            color = "red"
            folium.CircleMarker(location=coord,
                            radius=3,
                            fill=True,  # Set fill to True
                            tooltip=stop_name,
                            color=color,
                            fill_opacity=0.7).add_to(map_obj)
        # else:
            # color = "black"
    print('plotted points')
```

```python
conditional_stops_map = folium.Map(location=[28.628833, 77.206805], zoom_start=10)
conditionally_plot_points_on_map(conditional_stops_map, stops)
```

### plotting nearby stops

```python
def plot_nearby_points_on_map(map_obj, stops_df, query_coords, radius_km=2):
    for row in stops_df.itertuples():
        coord = (row.stop_lat, row.stop_lon)
        stop_distance = distance(query_coords, coord).km
        stop_name = row.stop_name
        if stop_distance <= radius_km:
            folium.CircleMarker(location=coord,
                                radius=3,
                                fill=True,  # Set fill to True
                                tooltip=stop_name,
                                color="green",
                                fill_opacity=0.7).add_to(map_obj)
        folium.CircleMarker(location=query_coords,
                                radius=8,
                                fill=True,  # Set fill to True
                                tooltip=stop_name,
                                color="blue",
                                fill_opacity=0.7).add_to(map_obj)
    print('plotted points')
```

```python
nearby_stops_map = folium.Map(location=[28.628833, 77.206805], zoom_start=10)
my_location = 28.544648392555725, 77.26840727835189
plot_nearby_points_on_map(nearby_stops_map, stops, my_location, radius_km=0.5)
```

```python
combined_gtfs_df = stops.merge(stop_times, on='stop_id')
combined_gtfs_df = combined_gtfs_df.merge(trips, on='trip_id')
combined_gtfs_df = combined_gtfs_df.merge(routes, on='route_id')
combined_gtfs_df.head(n=50)
```

```python
def plot_line_on_map(map_obj, route_name, route_df):
    points = list()
    for row in route_df.itertuples():
        coord = (row.stop_lat, row.stop_lon)
        points.append(coord)
    folium.PolyLine(points, color="red", tooltip=route_name).add_to(map_obj)
    # folium.PolyLine(points, color=random.choice(colors), tooltip=route_name).add_to(map_obj)
```

```python
def plot_route_on_map(map_obj, combined_df, route_name):
    try:
        route_trip = combined_gtfs_df[combined_gtfs_df['route_long_name'] == route_name].trip_id.values[0]
        route_stops = combined_gtfs_df[(combined_gtfs_df['route_long_name'] == route_name) & (combined_gtfs_df['trip_id'] == route_trip)].sort_values("stop_sequence")
        plot_points_on_map(map_obj, route_stops)
        plot_line_on_map(map_obj, route_name, route_stops)
    except IndexError as e:
        return pd.DataFrame()
```

```python
stop_A = 3127
stop_B = 230
```

```python
stop_A_name = stops[stops.stop_id == stop_A]  # finding stop names
stop_A_name
```

```python
stop_B_name = stops[stops.stop_id == stop_B]
stop_B_name
```

```python
# finding stop times with stop A in their trips
stop_A_stop_times = stop_times[stop_times.stop_id == stop_A]
stop_A_stop_times.head()
```

```python
stop_A_trips = stop_A_stop_times.trip_id
stop_A_trips
```

```python
trips.head()
```

```python
stop_A_trip_routes = trips[trips.trip_id.isin(stop_A_trips)].route_id.unique()
stop_A_trip_routes
```

```python
# finding stop times for stop A trips
# to know the stop sequence for all trips passing through stop A

stop_A_trips_stop_times = stop_times[stop_times.trip_id.isin(stop_A_trips)]
```

```python
# grouping stop A trip stop sequences / stop_times by trip_id

stop_A_trip_stop_times_group = stop_A_trips_stop_times.groupby("trip_id")
```

```python
stop_A_to_stop_B_trips = list()  # all trips from stop A to stop B
for trip_id, trip_stop_times in stop_A_trip_stop_times_group:  # iterating over grouped stop times
    # print(trip_id)
    if stop_B in trip_stop_times.stop_id.tolist():
        # check if stop A is before stop B in the sequence

        trip_stop_seq = trip_stop_times.stop_id.tolist()  # trips stop sequence
        stop_A_position = trip_stop_seq.index(stop_A)

        stop_B_position = trip_stop_seq.index(stop_B)
        print(trip_stop_seq)
        print("stop A position", stop_A_position)
        print("stop B position", stop_B_position)

        if stop_A_position < stop_B_position:
            stop_A_to_stop_B_trips.append(trip_id)
stop_A_to_stop_B_trips
```

```python
answer_route_ids = trips[trips.trip_id.isin(stop_A_to_stop_B_trips)].route_id
# answer_route_id
answer_route = routes[routes.route_id.isin(answer_route_ids)]
answer_route
```

## moving buses in "real-time"

```python
import time
from ipyleaflet import Map, Marker, AntPath
```

```python
realtime_data_dict = defaultdict(list)

for i in range(10):
    print(i)
    feed = gtfs_realtime_pb2.FeedMessage()
    response = requests.get(pb_url)

    feed.ParseFromString(response.content)
    for entity in feed.entity:
        vehicle_id = entity.vehicle.vehicle.id
        vehicle_lat = entity.vehicle.position.latitude
        vehicle_lon = entity.vehicle.position.longitude
        vehicle_route_id = int(float(entity.vehicle.trip.route_id))
        vehicle_timestamp = entity.vehicle.timestamp
        vehicle_trip_id = entity.vehicle.trip.trip_id

        vehicle_route_name = routes[routes.route_id == vehicle_route_id].route_long_name.squeeze()
        vehicle_datetime = datetime.fromtimestamp(vehicle_timestamp, ist)

        realtime_data_dict[vehicle_id].append((vehicle_lat, vehicle_lon))
    time.sleep(10)
```

```python
realtime_map = Map(center=[28.628833, 77.206805], zoom=10)

random_buses = random.sample(list(realtime_data_dict.keys()), 1)
random_buses = ["DL1PD3041", "DL1PD2849", "DL1PD0052"]
trackers = list()

for bus in random_buses:
    path = realtime_data_dict[bus]
    tracker = Marker(location=path[0])
    realtime_map.add_layer(tracker)
    trackers.append([bus, tracker])

    # AntPath to display the direction of the path:
    ant_path = AntPath(
        locations= path,
        dash_array=[1, 10],
        delay=1000,
        color='#7590ba',
        pulse_color='#3f6fba'
    )
    # add the AntPath layer:
    realtime_map.add_layer(ant_path)

# Simulate movement along the path:
display(realtime_map)
for bus, tracker in trackers:
    path = realtime_data_dict[bus]
    print(bus, path, tracker)
    for i in range(1, len(path)-1):
        tracker.location = path[i+1]  # update tracker to the next coordinate
```

```python
from time import sleep
from ipyleaflet import *

m = Map(center=(52, 10), zoom=8)

mark = Marker(location=(51.5, 10))
m.add_layer(mark)
```

```python
nb_steps = 10
step = 0.1
start_location = mark.location
end_location = (52.2, 10)

for i in range(nb_steps + 1):
    percentage_animation = i / nb_steps

    a = 1 - percentage_animation
    b = percentage_animation

    mark.location = (a * start_location[0] + b * end_location[0], a * start_location[1] + b * end_location[1])

    sleep(step)
```
